## Import Python Libraries

```python
import gspread
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
from datetime import datetime
from matplotlib.ticker import StrMethodFormatter, NullFormatter
from oauth2client.service_account import ServiceAccountCredentials

np.seterr(divide='ignore')
```

## Load Bitcoin Price Data from Google Drive

```python
try:
    # define the scope
    scope = ['https://spreadsheets.google.com/feeds','https://www.googleapis.com/auth/drive']
    # add credentials to the account
    creds = ServiceAccountCredentials.from_json_keyfile_name('client_secret.json', scope)
    # authorize the clientsheet 
    client = gspread.authorize(creds)
    # get the instance of the Spreadsheet
    sheet = client.open('Bitcoin Daily Price')
    # get the first sheet of the Spreadsheet
    sheet_instance = sheet.get_worksheet(0)
    # get all the records of the data
    records_data = sheet_instance.get_all_records()
    df = pd.DataFrame.from_dict(records_data)
except:
    df = pd.read_csv('../../github-bitcoin-scripts/bitcoin-data/bitcoin-daily-data.csv')
df.tail(1)
```

## Define Date Conversion Functions

```python
def str_to_datetime(strs):
    """Convert a list of "YYYY-MM-DD" strings to datetime objects."""
    dates = []
    for date in strs:
        if '-' in date:
            y, m, d = (int(x) for x in date.split("-"))
        elif '/' in date:
            m, d, y = int(date.split('/')[0]), int(date.split('/')[1]), \
            int(date.split('/')[2])+2000
        date = datetime(y, m, d)
        dates.append(date)
    return dates

def x_axis_dates(ax=None, fig=None):
    """Helper function to format the x axis as dates.

    Input:
    ax:  An Axes instance or an iterable of Axes instances.
    Optional, defaults to plt.gca()
    fig: The Figure instance containing those Axes or Axeses
    Optional, defaults to plt.gcf()
    """
    if ax is None: ax = plt.gca()
    if fig is None: fig = plt.gcf()
    loc = mdates.AutoDateLocator()
    fmt = mdates.AutoDateFormatter(loc)
    try:
        ax.xaxis.set_major_locator(loc)
        ax.xaxis.set_major_formatter(fmt)
    except AttributeError:
        for a in ax:
            # Fresh locators/formatters are needed for each instance
            loc = mdates.AutoDateLocator()
            fmt = mdates.AutoDateFormatter(loc)
            a.xaxis.set_major_locator(loc)
            a.xaxis.set_major_formatter(fmt)
    fig.autofmt_xdate()

orig_dates = df['date'].to_list()
datetimes = str_to_datetime(orig_dates)
dates = [mdates.date2num(date) for date in datetimes]
```

## Plot Bitcoin Daily Price Data (Sanity Check)

```python
prices = df['price'].to_list()
first_price = np.where(np.array(prices) > 0)[0][0]
nonzero_prices = prices[first_price:]
nonzero_dates = dates[first_price:]

ax = plt.gca()
plt.plot(nonzero_dates, nonzero_prices, color='#ff9900')
plt.yscale('log')
x_axis_dates()
ax.yaxis.set_major_formatter(StrMethodFormatter('${x:.2f}'))
ax.yaxis.set_minor_formatter(NullFormatter())
```

## Group Price Data by Month

```python
# Dates given in days since 1/1/1970
halving_dates = [(datetime(2012, 11, 28) - datetime(1970, 1, 1)).days,
                 (datetime(2016, 7, 9) - datetime(1970, 1, 1)).days,
                 (datetime(2020, 5, 11) - datetime(1970, 1, 1)).days,
                 (datetime(2024, 3, 26) - datetime(1970, 1, 1)).days]

zero_ind = len(df[df['price'] == 0])
xdata = [int(date) for date in dates[zero_ind:]]
ydata = list(df['price'][df['price'] > 0])

monthly_dates = []
for y in range(2009, 2022):
    for m in range(12):
        monthly_dates.append((datetime(y, m+1, 1) - datetime(1970, 1, 1)).days)
    
min_m_price = []
max_m_price = []
avg_m_price = []
plot_dates = []
days_until_next_halving = []
h = 0
for m in range(len(monthly_dates)-1):
    if (monthly_dates[m] in xdata) & (monthly_dates[m+1] in xdata):
        min_m_price.append(np.min(ydata[xdata.index(monthly_dates[m]):xdata.index(monthly_dates[m+1])]))
        max_m_price.append(np.max(ydata[xdata.index(monthly_dates[m]):xdata.index(monthly_dates[m+1])]))
        avg_m_price.append(np.mean(ydata[xdata.index(monthly_dates[m]):xdata.index(monthly_dates[m+1])]))
        plot_dates.append(np.mean(xdata[xdata.index(monthly_dates[m]):xdata.index(monthly_dates[m+1])]))
        while halving_dates[h] < plot_dates[-1]:
            h += 1
        days_until_next_halving.append(halving_dates[h] - plot_dates[-1])
    elif monthly_dates[m+1] in xdata:
        min_m_price.append(np.min(ydata[:xdata.index(monthly_dates[m+1])]))
        max_m_price.append(np.max(ydata[:xdata.index(monthly_dates[m+1])]))
        avg_m_price.append(np.mean(ydata[:xdata.index(monthly_dates[m+1])]))
        plot_dates.append(np.mean(xdata[:xdata.index(monthly_dates[m+1])]))
        while halving_dates[h] < plot_dates[-1]:
            h += 1
        days_until_next_halving.append(halving_dates[h] - plot_dates[-1])

ax = plt.gca()
plt.scatter(plot_dates, avg_m_price, c=days_until_next_halving, cmap='jet', vmin=0, vmax=1400)
plt.yscale('log')
x_axis_dates()
ax.yaxis.set_major_formatter(StrMethodFormatter('${x:.2f}'))
ax.yaxis.set_minor_formatter(NullFormatter())
```

```python
# Built to work on daily data, x-axis day-by-day counter starting at 1/1/1970
# 

def stock2fomo(x0, y0, a, b, xmin=14522, ymin=0.0001, xmax=20000, ymax=10**12):
    xfit = list(range(int(np.ceil(x0-1/b)), int(np.ceil(x0+1/b))))
    yfit = list(a*np.log((1+b*(np.array(xfit)-x0))/(1-b*(np.array(xfit)-x0)))+y0)
    indstart = 0
    indcurr = -1
    xshift = 0
    while indstart == 0:
        indcurr += 1
        if np.exp(yfit[indcurr]) > ymin:
            indstart = indcurr
    if indstart > 0:
        xfit = xfit[indstart:]
        yfit = yfit[indstart:]
        if xmin < xfit[0]:
            yfit = [np.log(ymin)]*int(xfit[0] - xmin) + yfit
            xfit = list(range(xmin, xfit[0])) + xfit
            xshift = xfit[0] - xmin
    return xfit, yfit, xshift


def leastsqrs_res(ydata, yfit):
    res = 0
    all_pos = True
    all_neg = True
    for i in range(len(ydata)):
        if ydata[i] - yfit[i] > 0:
            all_pos = False
        if ydata[i] - yfit[i] < 0:
            all_neg = False
        res += (ydata[i] - yfit[i])**2
    res = res/len(ydata)
    return res, all_pos, all_neg


def mk_s2fomo_df(s2fomo_df, xdata, ydata, x0_r, y0_r, a_r, b_r):
    starting_len = len(xdata)
    for x0 in x0_r:
        # round to nearest int
        x0_2 = int(round(x0, 0))
        for y0 in y0_r:
            # round to 2 decimal places
            y0_2 = round(y0, 2)
            for a in a_r:
                # round to 2 decimal places
                a_2 = round(a, 2)
                for b in b_r:
                    # round to 6 decimal places
                    b_2 = round(b, 6)
                    if len(s2fomo_df[(s2fomo_df['x0'] == x0_2) & (s2fomo_df['y0'] == y0_2) & (s2fomo_df['a'] == a_2) &
                                     (s2fomo_df['b'] == b_2)]) == 0:
                        xfit, yfit, xshift = stock2fomo(x0_2, y0_2, a_2, b_2, xmin=xdata[0],
                                                        ymin=np.exp(min(ydata))*0.1)
                        d_count = len(xdata) - xshift
                        if xdata[0] > xfit[0]:
                            start_ind = xfit.index(xdata[0])
                            xfit = xfit[start_ind:]
                            yfit = yfit[start_ind:]
                        elif xdata[0] < xfit[0]:
                            start_ind = xdata.index(xfit[0])
                            xdata = xdata[start_ind:]
                            ydata = ydata[start_ind:]
                        if len(xdata) < len(xfit):
                            xfit = xfit[:len(xdata)]
                            yfit = yfit[:len(xdata)]
                        elif len(xdata) > len(xfit):
                            xdata = xdata[:len(xfit)]
                            ydata = ydata[:len(xfit)]
                        res, all_pos, all_neg = leastsqrs_res(ydata, yfit)
                        row = pd.DataFrame([[x0_2, y0_2, a_2, b_2, res, all_pos, all_neg, d_count]],
                                           columns=['x0', 'y0', 'a', 'b', 'res', 'all_pos', 'all_neg', 'd_count'])
                        s2fomo_df = s2fomo_df.append(row, ignore_index=True)
    return s2fomo_df


def stock2fomochart(x0, y0, a, b):
    res = (1920/2.5, 1920/2.)
    size = res[0]/96.
    size_y = res[1]/96.

    # create figure and axis objects with subplots()
    fig = plt.subplots(figsize=(size, size_y), dpi=96)

    zero_ind = len(df[df['price'] == 0])
    xdata = [int(date) for date in dates[zero_ind:]]
    ydata = list(np.log(df['price'][df['price'] > 0]))
    plt.plot(xdata, ydata, color="#FF9900")

    # make a plot with different y-axis using second axis object
    xfit, yfit, xshift = stock2fomo(x0, y0, a, b, xmin=xdata[0], ymin=np.exp(min(ydata))*0.1)
#     print(len(xfit), len(yfit))
    plt.plot(xfit, yfit, color="black", linestyle='-', linewidth=0.75)

    x_axis_dates()
    plt.ylabel('BTCUSD', fontsize=15)

    old_xticks = plt.xticks()[0]
    new_xticks = []
    for t in range(len(old_xticks)):
        tick = old_xticks[t]
        new_xticks.append(tick)
        if t%4 == 0:
            new_xticks.append(tick+730)
        else:
            new_xticks.append(tick+731)        

    xticks_pos = [13879.0, 14609.0, 15339.0, 16070.0, 16800.0, 17531.0, 18261.0, 18992.0, 19722.0, 20453.0, 21183.0,
                  21914.0, 22644.0, 23375.0, 24105.0, 24836.0, 25566.0, 26297.0, 27027.0, 27758.0, 28488.0, 29219.0,
                  29950.0, 30681.0, 31411.0, 32142.0, 32872.0, 33603.0, 34333.0, 35064.0, 35794.0, 36525.0, 37255.0,
                  37986.0, 38716.0, 39447.0, 40177.0]
    xticks_labels = ["2008", "2010", "2012", "2014", "2016", "2018", "2020", "2022", "2024", "2026", "2028", "2030",
                     "2032", "2034", "2036", "2038", "2040", "2042", "2044", "2046", "2048", "2050", "2052", "2054",
                     "2056", "2058", "2060", "2062", "2064", "2066", "2068", "2070", "2072", "2074", "2076", "2078",
                     "2080"]
    xt_pos = []
    xt_lab = []
    x_ind = 0
    while (xticks_pos[x_ind] < xfit[-1]) & (x_ind < (len(xticks_pos)-1)):
        if xfit[-1] > 29219:
            x_ind += 1
            xt_pos.append(xticks_pos[x_ind])
            xt_lab.append(xticks_labels[x_ind])
            if x_ind < (len(xticks_pos)-2):
                x_ind += 2
            elif x_ind < (len(xticks_pos)-1):
                x_ind += 1
        elif xfit[-1] > 25566:
            x_ind += 1
            xt_pos.append(xticks_pos[x_ind])
            xt_lab.append(xticks_labels[x_ind])
            if x_ind < (len(xticks_pos)-1):
                x_ind += 1
        else:
            xt_pos.append(xticks_pos[x_ind])
            xt_lab.append(xticks_labels[x_ind])
            x_ind += 1
    if x_ind < len(xticks_pos):
        if xfit[-1] < 25566:
            xt_pos.append(xticks_pos[x_ind])
            xt_lab.append(xticks_labels[x_ind])
    plt.xticks(xt_pos, xt_lab, fontsize=14)
    if xfit[-1] > 25566:
        plt.yticks(np.log([0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000,
                           1000000000, 10000000000, 100000000000, 1000000000000, 10000000000000, 100000000000000]),
                   ["0.1¢", "1¢", "10¢", "$1", "$10", "$100", "$1k", "$10k", "$100k", "$1M", "$10M", "$100M", "$1B",
                    "$10B", "$100B", "$1T", "$10T", "$100T"], fontsize=14)
        plt.ylim(np.log(0.0001), np.log(1000000000000000))
    else:
        plt.yticks(np.log([0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000,
                           1000000000, 10000000000, 100000000000]),
                   ["0.1¢", "1¢", "10¢", "$1", "$10", "$100", "$1k", "$10k", "$100k", "$1M", "$10M", "$100M", "$1B",
                    "$10B", "$100B"], fontsize=14)
        plt.ylim(np.log(0.0001), np.log(1000000000000))
    plt.grid(alpha=0.5, linestyle='--', linewidth=0.75, axis='both')
    
    plt.title("Inverse Hyperbolic Tangent (Stock2FOMO) Model")

    plt.show()

#     save the plot as a file
#     fig.savefig('total_net_worth_usd_vs_btc.png')
```
